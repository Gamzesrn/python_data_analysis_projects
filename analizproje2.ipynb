```python
kb = df[df["Gender"] == "female"]["Height (cm)"]
eb = df[df["Gender"] == "male"]["Height (cm)"]
kadin_ortalama = kb.mean()
erkek_ortalama = eb.mean()
plt.hist(kb, label="Female")
plt.hist(eb, label="Male")
plt.title("Height Distribution of Persons by Gender")
plt.xlabel("Size")
plt.ylabel("Number")
plt.axvline(x=kadin_ortalama, color="blue", linestyle="--", label="Average of Female")
plt.axvline(x=erkek_ortalama, color="red", linestyle="--", label="Average of Male")
plt.legend()
```

```python
count = df.groupby(['Marital Status', 'Gender'])['Gender'].count().unstack()
married_count = count.loc['Married'].sum()
single_count = count.loc['Single'].sum()
plt.figure(figsize=(4, 4))
plt.bar(['Married', 'Single'], [married_count, single_count], width=0.3)
plt.title('Number of Married-Single')
plt.xlabel('Marital status')
plt.ylabel('Count')
```
